#### 📐 Lógica Matemática
Basado en la imagen técnica, el script utiliza las variables **$a$** (altura) y **$b$** (base) para ejecutar las siguientes fórmulas:
* **Perímetro:** $P = 2a + 2b$
* **Área:** $A = a \cdot b$

#### 💻 Estructura del Código
El software se organiza en tres pilares fundamentales:
1.  **Modularidad:** Uso de **funciones** independientes para el área y el perímetro, facilitando la lectura y el mantenimiento.
2.  **Interatividad:** Una función **`menu()`** centraliza las opciones, utilizando un ciclo **`while`** para que el usuario realice varios cálculos sin cerrar el programa.
3.  **Control de Flujo:** Estructuras **`if-elif-else`** que dirigen las decisiones del usuario hacia la operación correcta o hacia la salida del sistema.

Este enfoque modular enseña a separar la **interfaz** (menú) del **procesamiento** (fórmulas), una habilidad clave en el desarrollo de software.



```python
# Función para calcular el área: A = a * b
def calcular_area_rectangulo():
    print("\n--- Calcular Área ---")
    base = float(input("Ingresa la base (b): "))
    altura = float(input("Ingresa la altura (a): "))
    area = base * altura
    print(f"El área del rectángulo es: {area}")

# Función para calcular el perímetro: P = 2a + 2b
def calcular_perimetro_rectangulo():
    print("\n--- Calcular Perímetro ---")
    base = float(input("Ingresa la base (b): "))
    altura = float(input("Ingresa la altura (a): "))
    perimetro = 2 * altura + 2 * base
    print(f"El perímetro del rectángulo es: {perimetro}")

# Función principal del menú
def menu():
    continuar = True

    while continuar:
        print("\n============================")
        print("      MENÚ DE OPCIONES")
        print("============================")
        print("1. Calcular área rectángulo")
        print("2. Calcular perímetro rectángulo")
        print("3. Salir")

        opcion = input("\nSelecciona una opción (1-3): ")

        if opcion == "1":
            # Llamada a la función de área
            calcular_area_rectangulo()
        elif opcion == "2":
            # Llamada a la función de perímetro
            calcular_perimetro_rectangulo()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("Programa realisado por Oscar Flores NL = 19")
            # Cambiamos la variable a False para romper el ciclo while
            continuar = False
        else:
            # Por si el usuario escribe algo que no es 1, 2 o 3
            print("Opción no válida, intenta de nuevo.")

# Ejecución del programa
menu()
```
